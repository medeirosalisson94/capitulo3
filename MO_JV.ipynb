**1 - TRATAMENTO DE DADOS DE TEOR DE MATÉRIA ORGÂNICA - SERRA DE JOÃO DO VALE**

```python
#Importar bibliotecas
import scipy.stats as stats
import statsmodels
from statsmodels.stats.diagnostic import lilliefors
import pandas as pd

# Importar DataFrame
co2_pts_jv = pd.read_csv('C:\\Users\\Alisson Medeiros\\tese_cap6\\tabelas\\atributos_JV_mart.csv', sep = ';', encoding = 'UTF-8')
co2_pts_jv
#TESTE DE NORMALIDADE DAS AMOSTRAS DE CO2 NO SOLO - TESTE LILLIEFORS (COLMOGOROV-SMINORV)

#Teste de normalidade dos pontos amostrais
statsmodels.stats.diagnostic.lilliefors(co2_pts_jv.MO_g_Kg_1, dist = 'norm')

#O valor de p é  0.007403764916646883, MENOR que 0,05. NÃO É DISTRIBUIÇÃO NORMAL

#TESTE DE NORMALIDADE DAS AMOSTRAS DE CO2 NO SOLO - TESTE SCHAPIRO-WILK

stats.shapiro(co2_pts_jv['MO_g_Kg_1'])

#O valor de p é  0.02195657789707184, MENOR que 0,05. NÃO É DISTRIBUIÇÃO NORMAL

#TESTE DE NORMALIDADE DAS AMOSTRAS DE CO2 NO SOLO - TESTE ANDERSON-DARLING

stats.anderson(co2_pts_jv['MO_g_Kg_1'],'norm')

#O valor (statistic=1.0121881156867865, critical_values=array([0.517, 0.589, 0.707, 0.824, 0.98 ]). É DISTRIBUIÇÃO NORMAL

#TESTE PARA DADOS NÃO PARAMÉTRICOS - KRUSKAL-WALLIS
import numpy as np
from scipy.stats import kruskal

kruskal(*co2_pts_jv)

#Valor de p do teste de Kruskal-Wallis: 0.36787944117144245 > 0,05. NÃO HÁ DIFERENÇA ESTATISTICAMENTE SIGNIFICATIVA AS AMOSTRAS.
```
